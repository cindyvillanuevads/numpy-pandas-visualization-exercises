```python
import matplotlib.pyplot as plt  
```

```python
import math
# y=0n+1


plt.figure(figsize = (10,10))

x = range (1,5)
y1 = [1 for n in x]
y2 = [ math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [n **2 for n in x]
y6 = [ 2**n for n in x]
y7= [( math.factorial(n))* .02 for n in x]
y8= [(n**n) *.05 for n in x]




plt.plot(x,y1, ls = '-', marker = '.', label = '$O(1)$', color = "m")
plt.plot(x,y2, ls ="--", marker = "*", label = "$O(log n)$")
plt.plot(x,y3, ls = '-', marker = 'X', label = '$0(n)$', color = 'g')
plt.plot(x,y4, ls = '-.', marker = 'X', label = r'$O(n log n)$', color = 'orange')
plt.plot(x,y5, ls = '-', marker = '*', label = r'$O(n^2)$', color = 'pink')
plt.plot(x,y6, ls = '-', marker = '+', label = r'$O(2^n)$', color = 'purple')
plt.plot(x,y7, ls = '-', marker = '+', label = r'$O(n!)$', color = 'red')
plt.plot(x,y8, ls = '-', marker = 'X', label = r'$O(n^n)$', color = 'darkgreen')


plt.grid(True, ls = '--', color = 'gray',)
plt.title('Big O Notation', fontsize=30, fontweight='bold',
         bbox={'facecolor': 'lightblue', 'alpha': 0.5, 'pad': 5})

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.show()
```
