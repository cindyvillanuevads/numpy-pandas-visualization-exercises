### 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.



```python
def get_db_url(username,password,host,db_name):
    return f'mysql+pymysql://{username}:{password}@{host}/{db_name}'
    
from env import host, username, password
```

```python
url2 =  get_db_url(username,password,host,'chipotle')
```

```python
chipotle_df = pd.read_sql('SELECT * FROM orders', url2)
```

```python
chipotle_df.head()
```

```python
chipotle_df.groupby("item_name").sum(["quantity","item_price"])
```

```python
chipotle_df['item_price'] = chipotle_df['item_price'].str.replace('$', '').astype('float')
```

```python
chipotle_df
```

```python
popular = chipotle_df.groupby("item_name").item_price.sum().nlargest(4, keep="all")
```

```python
popular
```

```python
popular.plot(kind = "bar")
```
