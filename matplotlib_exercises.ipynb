1. Use matplotlib to plot the following equation:

y = (x**2) - x +2 

```python
import matplotlib.pyplot as plt
```

```python
print(plt)
```

```python
x = range(-50, 51)
y = [ (n**2) - n + 2 for n in x]
plt.plot(x,y)
plt.annotate('The origin', xy = (0, 0), xytext = (30, 0), arrowprops = dict(facecolor = 'red'))
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    * y = √x
    * y =x **3
    * y=2**x
    * y=1/(x+1)

```python
# y = √x
import math
x = range (0, 12)
y = [math.sqrt(n) for n in x] 
plt.plot(x,y)
```

```python
# y =x **3
x = range (0, 10)
y = [n ** 3 for n in x]
plt.plot(x,y)
```

```python
# y=2**x
x = range (0, 8)
y = [2 ** n for n in x]
plt.plot(x,y)
```

```python
# y= 1/(x+1)
x = range (0,11)
y = [1 / (n + 1) for n in x]
plt.plot(x,y)
```

3. Combine the figures you created in the last step into one large figure with 4 subplots

```python
x = [x for x in range(0,11)]
# y = √x
y1 = [math.sqrt(n) for n in x] 
# y =x **3
y2 = [n ** 3 for n in x]
# y=2**x
y3 = [2 ** n for n in x]
# y= 1/(x+1)
y4 = [1 / (n + 1) for n in x]
```

```python
# Define a figure
plt.figure(figsize = (10, 8))


plt.subplot(2,2,1)
plt.plot(x,y1, ls = '-', marker = '*', color = 'black')
plt.title('$y = √x$')
plt.grid(True, ls = '--', color = 'gray')

plt.subplot(2,2,2)
plt.plot(x,y2, ls = '-', marker = '+', color = 'black')
plt.title('$y = x **3$')
plt.grid(True, ls = '--', color = 'gray')

plt.subplot(2,2,3)
plt.plot(x,y3, ls = '-', marker = '*', color = 'black')
plt.title('$y = 2**x$')
plt.grid(True, ls = '--', color = 'gray')

plt.subplot(2,2,4)
plt.plot(x,y3, ls = '-', marker = '*', color = 'black')
plt.title('$y = 1/(x+1)$')
plt.grid(True, ls = '--', color = 'gray')

# add padding to avoid overlap of subplots
plt.tight_layout()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# Define a figure
plt.figure(figsize = (10, 8))
plt.suptitle(' Combined Figures', fontsize = 16)

plt.subplot(2,2,1)
plt.plot(x,y1, ls = '-', marker = '*', color = 'blue',label = r'$y = \sqrt{x}$')
plt.title(r'$y = \sqrt{x}$')
plt.grid(True, ls = '--', color = 'gray')
plt.legend()



plt.subplot(2,2,2)
plt.plot(x,y2, ls = '-', marker = '+', color = 'green', label = r'$y = x ^3$')
plt.title(r'$y = x ^3$')
plt.grid(True, ls = '--', color = 'gray')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x,y3, ls = '-', marker = '*', color = 'orange', label = r'$y = 2^x$')
plt.title(r'$y = 2^x$')
plt.grid(True, ls = '--', color = 'gray')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x,y3, ls = '-', marker = '*', color = 'm', label = r'$y = \frac{1}{x+1}$' )
plt.title(r'$y = \frac{1}{x+1}$')
plt.grid(True, ls = '--', color = 'gray')
plt.legend()


# add padding to avoid overlap of subplots
plt.tight_layout()
```
